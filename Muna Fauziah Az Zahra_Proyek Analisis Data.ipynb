Cek Outlier

```python
# Menggunakan dataset sebelumnya (dari gambar/deskripsi "order_items_df")

# Menghitung IQR dan batas outlier pada kolom payment_value
Q1_real = order_items_df['payment_value'].quantile(0.25)
Q3_real = order_items_df['payment_value'].quantile(0.75)
IQR_real = Q3_real - Q1_real

# Batas bawah dan atas
lower_bound_real = Q1_real - 1.5 * IQR_real
upper_bound_real = Q3_real + 1.5 * IQR_real

# Mendeteksi outlier
outliers_real = order_items_df[
    (order_items_df['payment_value'] < lower_bound_real) |
    (order_items_df['payment_value'] > upper_bound_real)
]

# Dataset tanpa outlier
order_items_df_cleaned = order_items_df[
    (order_items_df['payment_value'] >= lower_bound_real) &
    (order_items_df['payment_value'] <= upper_bound_real)
]

(Q1_real, Q3_real, IQR_real, lower_bound_real, upper_bound_real), outliers_real.shape, order_items_df_cleaned.shape
```

```python
order_items_df
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Resample untuk mendapatkan data tahunan dan total payment value (revenue)
years_orders_df = order_items_df.resample(rule='Y', on='order_purchase_timestamp').agg({
    "order_id": "nunique",
    "payment_value": "sum"
})

# Ubah format tanggal menjadi tahun-bulan dan assign ke kolom baru
years_orders_df['year_month'] = years_orders_df.index.strftime('%Y-%m')  # Changed to 'year_month'

# Reset index dan rename kolom
years_orders_df = years_orders_df.reset_index()
years_orders_df.rename(columns={
    "order_id": "order_count",
    "payment_value": "revenue"
}, inplace=True)

# Plot line chart untuk total payment value (revenue)
plt.figure(figsize=(10, 5))
plt.plot(years_orders_df["year_month"], years_orders_df["revenue"], marker='o', linewidth=2, color="#72BCD4", label="Payment Value") # Changed to 'revenue'

# Menambahkan trendline
x = np.arange(len(years_orders_df))
y = years_orders_df["revenue"].values  # Changed to 'revenue' instead of 'order_count'
coefficients = np.polyfit(x, y, 1)
poly = np.poly1d(coefficients)
trendline = poly(x)

plt.plot(years_orders_df["year_month"], trendline, color='orange', linestyle='--', label="Trendline") # Changed to 'year_month'

# Menambah judul dan label
plt.title("Total Payment Value by Year with Trendline", fontsize=20)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Payment Value (Revenue)", fontsize=12)

# Menampilkan legenda
plt.legend()

# Menampilkan plot
plt.show()
```

## Analisis Lanjutan (RFM)

```python
# Konversi tanggal
order_items_df["order_purchase_timestamp"] = pd.to_datetime(order_items_df["order_purchase_timestamp"])
current_date = pd.to_datetime("2024-12-28")  # Tanggal referensi
```

```python
order_items_df = pd.merge(order_items_df, orders_df[['order_id', 'customer_id']], on='order_id', how='left')
```

```python
# Hitung RFM
rfm = order_items_df.groupby('customer_id').agg({
    'order_purchase_timestamp': lambda x: (current_date - x.max()).days,  # Recency
    'order_id': 'nunique',  # Frequency
    'payment_value': 'sum'  # Monetary
})

# Ubah nama kolom
rfm.rename(columns={
    'order_purchase_timestamp': 'Recency',
    'order_id': 'Frequency',
    'payment_value': 'Monetary'
}, inplace=True)
```

```python
order_items_df
```

```python
# Membuat figure dan axes untuk 3 plot secara sejajar
fig, axes = plt.subplots(1, 3, figsize=(20, 6))  # 1 baris, 3 kolom

# Histogram Recency
sns.histplot(rfm['Recency'], bins=20, kde=True, color="#72BCD4", ax=axes[0])
axes[0].set_title('Distribution of Recency')
axes[0].set_xlabel('Recency (days)')
axes[0].set_ylabel('Frequency')

# Histogram Frequency
sns.histplot(rfm['Frequency'], bins=20, kde=True, color="#72BCD4", ax=axes[1])
axes[1].set_title('Distribution of Frequency')
axes[1].set_xlabel('Frequency')
axes[1].set_ylabel('Frequency')

# Histogram Monetary
sns.histplot(rfm['Monetary'], bins=20, kde=True, color="#72BCD4", ax=axes[2])
axes[2].set_title('Distribution of Monetary')
axes[2].set_xlabel('Monetary Value')
axes[2].set_ylabel('Frequency')

# Menampilkan plot
plt.tight_layout()  # Mengatur layout agar tidak tumpang tindih
plt.show()
```
